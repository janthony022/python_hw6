```python
# Fill each persons name and twitter handle, using groups, should look like:
    
# ==============
# Full Name/Twitter
# ==============
# Derek Hawkins/@derekhawkins
# Erik Sven-Osterberg/@sverik
# Ryuan Butz/@ryanbutz
# Example Exampleson/@example
# Ripal Pael/@ripalp
# Darth Vader/@darthvader

# Hint: use .group()

import re
with open('names.txt') as f:
    data = f.readlines()

pattern = re.compile("([A-Z][a-z]+), ([\w-]*)([A-Z][a-z]+).*(@[a-zA-Z0-9]+$)")

for person in data: 
    match = pattern.search(person) 
    
    if match: 
        print('\n', f'{match.group(3)} {match.group(2)}{match.group(1)} / {match.group(4)}')
        
```

```python
# Use python to read the file regex.test.txt and print the last name on each line using
# regular expressons and groups(return None for names with no first and last name, or names
# that aren't properly capitalized)
                              
# Hint: use with open() and readlines()

with open('regex_test.txt') as f:
    data = f.readlines()
    
    for name in data:
        validate_name = re.compile("(^[A-Z][a-z]\w+\s[A-Z]\s?[A-Za-z]\w+\s?[A-Za-z]\w+)")
        match = validate_name.search(name)
        
        if match:
            print(match.group(1))
        else:
            print("None")
```
